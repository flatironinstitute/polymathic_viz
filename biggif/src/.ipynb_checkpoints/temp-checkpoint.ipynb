```python
import logging
import importlib
import os
import polymathic_viz as p
import time

def main():
    start_time = time.time()
    importlib.reload(p)
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
    
    total_num_gifs =  0
    
    # process goes here
    num_datasets = len(p.datasets)
    for i in range(num_datasets):
        dataset_name = os.path.basename(p.datasets[i])
        config_data = p.get_dataset_config(dataset_name, p.config_dir)
        
        # loop over all files
        files = config_data.get("file_names")
        numfiles = len(files)
        for j in range(len(files)):
            # loop over t0 fields
            numt0 = len(config_data.get("t0"))
            numt1 = len(config_data.get("t1"))
            # loop over t2 fields
            numt2 = len(config_data.get("t2"))
            numt2 = numt2 * 2
            break 
        print(f"{dataset_name} has {numfiles} files")
        print(f"{dataset_name} has {numfiles*(numt0 + numt1 + numt2)} gifs (if only visualize initial condition 0 in each field)")
        total_num_gifs += numfiles*(numt0 + numt1 + numt2)
    print(f"total number of gifs: {total_num_gifs}")
        
    end_time = time.time()
    duration = end_time - start_time
    logging.info(f"Main Completed! Time taken: {duration:.4f} seconds.")
    
    
    
if __name__ == "__main__":
    main()
```
