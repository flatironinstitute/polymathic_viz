## 1. Create Viewable Gifs and save to "collection" folder

Some notes:
* `print_datasets()` - prints the dataset list I have stored
* `read_file_as_array()` - converts the specified field into a colored ready to visualize numpy array
    * Input: `dataset_num` (required) - dataset_num in the dataset list
        * `f` (optional) - file number (0 by default)
        * `t` (optional) - t()_field (0 by default)
        * `k` (optional) - field number (-1 the last field by default)
        * `c` (optional) - initial condition number (0 by default)
    * Output: `array`, `names` 
        * for 2D: `array` has shape (ts, h, w, 3)
        * for 3D: `array` has shape (num_slices, ts, h, w, 3)
        * `names` is used for writing gif (following my naming convention)
            ```
            <dataset_name>-f<file_num>-<field_name>-s<slice_num_for_3D>-c<initial_condition>
            ```
* `write_gif()` - saves the .gif file for viewing and the .npy (array) file for further processing
    * Input: `array`, `names` (required)
        * `save_to_dir` (optional) - custom destination directory (`output_dir` by default)
        * `fps` (optional) - custom frames per second (7 by default) 
    * Output: None
    

````python
from polymathic_viz import print_datasets, read_file_as_array, write_gif
````

````python
print_datasets()
````

````python
array, names = read_file_as_array(1)
write_gif(array, names, save_to_dir="../raw")
````

## 3. Postprocessing Gifs into "processed" folder

Some notes:

After forming the "unprocessed" directory of all the gifs to put into the collage, you can process each .npy file info the right height, width, and timestamps to arrange into the biggif.
* `postprocess_next()` - this function processes the next gif (processed gifs one-by-one) so you can make sure you want to save
    * Inputs:
        * `from_dir` - the unprocessed directory (default: "../unprocessed")
        * `to_dir` - the destination directory (default: "../processed")
        * `mode` - the mode under which to run this function in (`"r"` by default)
            * `"r"` - read mode, only displays the cropped post processed results
            * `"w"` - write mode, saves the post processed results
            * `"x"` - remove mode, removes the unprocessed version to release its memory
        * `sh` - square image height in the collage (`256` by default)
        * `sw` - square image width in the collage (`256` by default)
        * `sl` - standard length/time stamps of images in the colalge (`200` by default)
    * Output:
        * `None`, just saves the processed gif into `to_dir`

Note: all images are either cropped or expanded to fit `sl` at "write-time" (mode "w") only


````python
from polymathic_viz import postprocess_next
````

````python
# observe the next gif i am post processing
postprocess_next(mode="r")
````

````python
# looks good!!! I am ready to postprocess it and remove the unprocessed one
postprocess_next(mode="rwx")
````

## 4. Create the final Biggif Collage and save to "output" folder

Some notes:

Once the `processed` folder is populated with enough gifs, you can piece together the biggif. Piecing together an aesthetic biggif requires control, so I implemented a separate `collage_maker` class to do the trick.
* `collage_maker()` (constructor)
    * Input: `numRows`,`numColumns` (required)
        * `ih` - individual image height (`256` by default)
        * `iw` - individual image width (`256` by default)
        * `npy_dir` - where to look for the npy files (`../processed` by default)


The control loop is going to repeatedly manually run two instance functions:
* `show_next()` - show the next gif that I can put in
    * Input: `shape` - the shape of the gif (optional)
        * `"s"` for square gifs (default)
        * anything else (typically `"r"`) for rectangular gifs
* `put()` - put the current displayed selection into the biggif if you are satisfied with it
* `save()`- saves two things:
    1. the gif collage as mp4 file
    2. `collage_sequence` - a numpy array of paths to the individual gifs in the collage in sequence (for recreating and editing later)
    * Input:
        * `to_dir` - output directory ("../output" by default)
        * `save_as` - file name ("final.mp4" by default)
        * `fps` - frames per second of the gifs (`7` by default)

Note: the constructor won't create the object and will throw `ValueError` if there is not enough npy files to fill the collage in teh `npy_dir` directory

Note: make sure you like it before you call `put()`, or else the only way to go back and edit it is through the saved `collage_sequence` npy file.




````python
from polymathic_viz import collage_maker
````

````python
# create an instance of collage_maker
c = collage_maker(1, 1)
````

````python
# observe the next option
c.show_next(shape="s")
````

````python
# once satisfied, put this gif into the biggif
c.put()
````

````python
# save the biggif
c.save()
````

## 5. Recreate the final Biggif Collage for further editing (if needed)

Some notes: 

You may need to recreate the final gif collage for many reasons including editing the order of the gifs, flipping the orientation of the gifs, etc. You may do this through accessing the saved `collage_sequence` npy file along with the mp4.

* recreate()
    * Inputs:
        * source - the path to the `collage_sequence` npy file (required)
        * save_as - save the recreated biggif as "*.mp4" (required)
        * numRows - number of rows of the new biggif (required)
        * numCols - number of columns of the new biggif (required) 
        * mode - mode to run recreate under
            * `"d"` - display mode, just displays the recreated biggif (default)
            * `"s"` - save mode, saves the recreated biggif

        * fps - frames per second (`7` by default)
        * compress - skip every how many frames (`1` by default which means does not skip frame, `2` skips every other frame, and vice versa)
        * ih - image height (`256` by default)
        * iw - image width (`256` by default)

Note: you can edit the saved npy `collage_sequence` to change the biggif appearance to modify things such as placement of gifs

````python
from polymathic_viz import recreate
recreate(source="../output/final.npy", save_as="../output/final_copy.npy", numRows = 1, numCols = 1, mode="d")
````

````python
# display the recreated gif and also save it
recreate(source="../output/final.npy", save_as="../output/final_copy.mp4", numRows = 1, numCols = 1, mode="ds")
````
